```python
# prompt: if absence_days is less then equal to 3 replace with 1, if more then 3 and greater than equal 6 replace with 2 if the  absence_days is less then 10 replace with 3 else 4

import numpy as np
def categorize_absences(x):
    if x <= 3:
        return 1
    elif 3 < x <= 6:
        return 2
    elif 6 < x < 10:
        return 3
    else:
        return 4

df['category'] = df['absence_days'].apply(categorize_absences)
print(df)
```

```python
# prompt: one hot encode category ( to 4 columns 1,2,3,4), drop absence_days

import pandas as pd
df = pd.get_dummies(df, columns=['category'], prefix='category_absence_days')
df.drop('absence_days', axis=1, inplace=True)
df
```

```python
def percentage_score(x):
    if x <= 60:
        return 1
    elif 60 < x <= 70:
        return 2
    elif 70 < x < 80:
        return 3
    elif 80 < x < 90:
        return 4
    else:
        return 5

df['category_percentage_score'] = df['percentage_score'].apply(percentage_score)
print(df)
```

```python
# prompt: percentage_score drop

df.drop('percentage_score', axis=1, inplace=True)
print(df)
```

```python
# prompt: change true and false to 1 and 0 incategory_absence_days_1	category_absence_days_2	category_absence_days_3	category_absence_days_4

df['category_absence_days_1'].replace([True, False], [1, 0], inplace=True)
df['category_absence_days_2'].replace([True, False], [1, 0], inplace=True)
df['category_absence_days_3'].replace([True, False], [1, 0], inplace=True)
df['category_absence_days_4'].replace([True, False], [1, 0], inplace=True)
```

```python
def weekly_self_study_hours(x):
    if x <= 10:
        return 1
    elif 10 < x <= 20:
        return 2
    elif 20 < x <=30:
        return 3
    elif 30 < x <=40:
        return 4
    else:
        return 5

df['category_weekly_self_study_hours'] = df['weekly_self_study_hours'].apply(weekly_self_study_hours)
print(df)
```

```python
# prompt: data frame should contain following fields-'part_time_job', 'extracurricular_activities', 'career_aspiration',
#        'math_score', 'history_score', 'physics_score', 'chemistry_score',
#        'biology_score', 'english_score', 'geography_score', 'total_score',
#        'category_absence_days_1', 'category_absence_days_2',
#        'category_absence_days_3', 'category_absence_days_4',
#        'category_p
X= df[['part_time_job', 'extracurricular_activities',
       'category_absence_days_1', 'category_absence_days_2',
       'category_absence_days_3', 'category_absence_days_4',
        'category_weekly_self_study_hours_1',
       'category_weekly_self_study_hours_2', 'category_weekly_self_study_hours_3',
       'category_weekly_self_study_hours_4', 'category_weekly_self_study_hours_5']]


y=df[['category_percentage_score']]
```

```python
# prompt: train,test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
def tranform_to(a):
  b = np.zeros((len(a), 5))
  for k in range(len(a)):
    if a[k]==1:
      b[k][0]=1
    elif a[k]==2:
      b[k][1]=1
    elif a[k]==3:
      b[k][2]=1
    elif a[k]==4:
      b[k][3]=1
    elif a[k]==5:
      b[k][4]=1
  return b
```

```python
from keras.models import Sequential
from keras.layers import Dense,Flatten

model = Sequential()

model.add(Dense(5, activation='sigmoid', input_shape=(11,)))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python
import numpy as np
y_train_as_np = np.array(y_train)

y_train_f=tranform_to(y_train_as_np)
model.fit(X_train,y_train_f , epochs=25)
```

```python
# prompt: test the model using accuracy_score

import numpy as np
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
y_test_as_np = np.array(y_test)
y_test_f=tranform_to(y_test_as_np)

max_values = np.max(y_pred, axis=1)
y_pred = np.where(y_pred == max_values[:, np.newaxis], 1, 0)

print(accuracy_score(y_test_f, y_pred))

```

```python
import pickle
from fastapi import FastAPI, HTTPException
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
import numpy as np
from colabcode import ColabCode

# Define request body schema using Pydantic
class PredictionData(BaseModel):
    part_time_job: int
    extracurricular_activities: int
    absence_days: int
    weekly_self_study_hours: int

# Load your pickled model
with open('model.pkl', 'rb') as f:
    model = pickle.load(f)

def part_time_job(x):
    if x==0:
        return [0]
    else:
        return [1]

def extracurricular_activities(x):
    if x==0:
        return [0]
    else:
        return [1]

# Categorize absence days
def categorize_absences(x):
    if x <= 3:
        return [1,0,0,0]
    elif 3 < x <= 6:
        return [0,1,0,0]
    elif 6 < x < 10:
        return [0,0,1,0]
    else:
        return [0,0,0,1]

# Categorize weekly self-study hours
def weekly_self_study_hours(x):
    if x <= 10:
        return [1,0,0,0,0]
    elif 10 < x <= 20:
        return [0,1,0,0,0]
    elif 20 < x <= 30:
        return [0,0,1,0,0]
    elif 30 < x <= 40:
        return [0,0,0,1,0]
    else:
        return [0,0,0,0,1]

# Create an instance of FastAPI
app = FastAPI()

origins = [
    "http://localhost.tiangolo.com",
    "https://localhost.tiangolo.com",
    "http://localhost",
    "http://localhost:3001",
    "http://localhost:8080",
]

app.add_middleware(
    CORSMiddleware,
    allow_origins=origins,
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

@app.get("/")
def test():
  return f"Hello World"

# Endpoint for making predictions
@app.post("/predict/")
async def predict(data: PredictionData):
    try:
        # Prepare input features for prediction
        part_time_job_array = part_time_job(data.part_time_job)
        extracurricular_activities_array = extracurricular_activities(data.extracurricular_activities)
        absence_days_array = categorize_absences(data.absence_days)
        weekly_self_study_hours_array = weekly_self_study_hours(data.weekly_self_study_hours)

        # Concatenate the arrays
        X = np.concatenate((
            part_time_job_array,
            extracurricular_activities_array,
            absence_days_array,  # Ensure 2D array
            weekly_self_study_hours_array # Ensure 2D array
        ))
        X = X.reshape((1, 11))

        # Make prediction
        prediction = model.predict(X)

        # Convert predicted probabilities to binary values
        max_values = np.max(prediction, axis=1)
        prediction_binary = np.where(prediction == max_values[:, np.newaxis], 1, 0)
        prediction = prediction_binary.tolist()
        index_of_1 = prediction[0].index(1)
        print("Index of 1:", index_of_1)
        return {"prediction": index_of_1}
    except Exception as e:
        # Log the error with detailed information
        import traceback
        traceback.print_exc()
        # Return an HTTP 500 error response with a detailed error message
        raise HTTPException(status_code=500, detail=f"An error occurred during prediction: {str(e)}")

# Run the FastAPI app with ColabCode
cc = ColabCode(port=12000, code=False)
cc.run_app(app=app)
```
